```python
"""9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into
suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where
size='Varies with device')
"""
#  'Size' in side the total number of  'Varies with devise'
df[df['Size'] == 'Varies with device'].count()
```

```python
# Drop entries where 'Size' is 'Varies with device'
df = df[df['Size'] != 'Varies with device']
```

```python
# Replace 'M' with 'e6', 'K' with 'e3', and convert to numeric
df['Size'] = df['Size'].replace({'M': '000000', 'k': '000'}, regex=True)
```

```python
#replace the "." into ""(No space)
df['Size'] = df['Size'].str.replace('.' ,'')
```
